```python
# Load data from netlogo's behaviorspace "spreadsheet" output
# ** DIFFERENT THAN THE LOADERS IN OTHER NOTEBOOKS: needs to parse netlogo array
def loadBehaviorspaceAsStrings(filename):
    lines = []
    with open(filename, 'rb') as csvfile:
         reader = csv.reader(csvfile, delimiter=',', quotechar='"')
         for row in reader:
             lines.append(row)

    metadata = lines[0:5]
    
    _data = lines[6:]
    _data = np.array(lines[6:], dtype=object).T
    for i in range(1, len(_data)):
        for j in range(1, len(_data[i])):
            try:
                _data[i][j] = _data[i][j]
            except:
                continue
    data = {}
    for col in _data:
        data[col[0]] = np.array(col[1:], dtype=str)
            
    return (metadata, data)

def parseNetlogoArray(s):
    return eval(s.replace(' ', ', '))
```

```python
def kl(p, q):
    epsilon = 0.0001 # avoid log(0)
    psum = sum(p)*1.0
    qsum = sum(q)*1.0
    pn = [epsilon + pp/psum for pp in p]
    qn = [epsilon + qq/qsum for qq in q]
    ii = [0 if pn[i]==0 else pn[i] * math.log(qn[i]/pn[i]) for i in range(len(p))]
    return -sum(ii)

kl([.68, .10, .22], [.66,.04,.30])

def klcounts(i, j):
    return [counts[i][0], counts[j][0], kl(counts[i][1], counts[j][1])]
```
